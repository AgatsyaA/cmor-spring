📌 Overview

This project introduces the concept of a Multi-Layer Neural Network, also known as a Multi-Layer Perceptron (MLP), a powerful machine learning model capable of learning complex, non-linear relationships in data.

The model consists of multiple layers of interconnected neurons that transform input data into meaningful predictions through learned weights.

Multi-Layer Perceptron (MLP) consists of fully connected dense layers that transform input data from one dimension to another. It is called multi-layer because it contains an input layer, one or more hidden layers and an output layer. The purpose of an MLP is to model complex relationships between inputs and outputs.

* Input Layer: Each neuron or node in this layer corresponds to an input feature. For instance, if you have three input features the input layer will have three neurons.
* Hidden Layers: MLP can have any number of hidden layers with each layer containing any number of nodes. These layers process the information received from the input layer.
* Output Layer: The output layer generates the final prediction or result. If there are multiple outputs, the output layer will have a corresponding number of neurons.

🧠 What is a Multi-Layer Perceptron?

A Multi-Layer Perceptron (MLP) is a type of artificial neural network composed of multiple layers of neurons.

👉 In simple terms:
It learns patterns by passing data through multiple layers, extracting deeper insights at each step.


---
🎯 Objective
* Model complex, non-linear relationships
* Improve predictive performance over simple models

⚙️ Methodology / Concept Explanation

An MLP consists of three main components:

🔹 1. Input Layer

Receives raw input features

🔹 2. Hidden Layers

Perform transformations using:
* Weights
* Biases
* Activation functions (e.g., ReLU)

🔹 3. Output Layer

Produces final predictions

🔹 Learning Process

* Forward Pass

  Data flows from input → hidden → output

* Loss Calculation

  Measures error between predicted and actual values
* Backpropagation

  Gradients are computed and propagated backward

* Weight Updates

  Optimizer (e.g., Adam) updates weights to reduce error

📌 Conclusion

Multi-Layer Perceptrons are powerful neural network models capable of learning complex, non-linear relationships in data. By leveraging multiple layers and iterative optimization, they provide high predictive accuracy and are widely used in advanced machine learning and deep learning applications.

In this project, an MLP was developed to classify wines using chemical features, covering data preprocessing, training, evaluation, and tuning. The baseline model achieved strong performance (0.944 accuracy), and tuning did not improve accuracy, indicating the baseline was already near-optimal for this dataset. However, tuning provided valuable insights into architecture, regularization, and learning rates. Loss curve analysis showed smooth and stable convergence, with no signs of overfitting or optimization issues.
