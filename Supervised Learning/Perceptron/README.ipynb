📌 Overview

This project introduces the Perceptron, the simplest form of an artificial neural network and a fundamental building block for more complex models.

It is a single-layer binary classifier that learns to separate data using a linear decision boundary.

A Perceptron is the simplest form of a neural network that makes decisions by combining inputs with weights and applying an activation function. It is mainly used for binary classification problems. It forms the basic building block of many deep learning models.

* Takes multiple inputs and assigns weights
* Computes a weighted sum and applies a threshold
* Outputs either 0 or 1 (binary outcome)
* Forms the foundation of larger neural networks

🧠 What is a Perceptron?

A Perceptron is a supervised learning algorithm used for binary classification.

👉 In simple terms:
It draws a straight line (decision boundary) to separate two classes.

1. Inputs: These are the features or measurable attributes of a data point that the perceptron uses to make a decision. Each input provides a signal that contributes to the final output.

2. Weights: Weights determine how strongly each input contributes to the prediction. A larger weight means the corresponding input has a higher impact.

3. Bias: The bias is a constant value added to the weighted sum to shift the decision boundary.

4. Net Input (Weighted Sum): This is the combined effect of all inputs and their weights.

5. Activation Function (Step Function): The activation function converts the numerical input into a binary output.

⚙️ Methodology / Concept Explanation

The perceptron works using a linear function:

y=sign(w⋅x+b)

🔹 Working Steps:
* Input Processing
  
  Multiply inputs by weights
* Add Bias
  
  Shift decision boundary
* Activation Function
  
  Apply step (sign) function
* Output → 0 or 1
  
  Weight Update Rule
* Adjust weights based on errors
  
  Improves classification over time

The perceptron is a foundational machine learning model that demonstrates how simple linear classifiers can learn from data.
