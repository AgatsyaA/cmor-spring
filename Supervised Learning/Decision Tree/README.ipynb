📌 Overview: This project demonstrates the implementation of a Decision Tree Classifier to predict whether a user will click on an online advertisement. The model learns decision rules from input features and classifies users based on their behavior patterns.

🧠 What is a Decision Tree?

A Decision Tree is a supervised machine learning algorithm that uses a tree-like structure to make decisions. It splits data based on feature conditions, forming branches that lead to final predictions at leaf nodes.

<img src= 'https://media.geeksforgeeks.org/wp-content/uploads/20250626155553066708/_what_is_a_decision_tree_.webp'>

It is widely used for:
* Classification (categorical output)
* Regression (continuous output)

🎯 Problem Statement

The objective of this project is to:
*   Predict whether a user clicks on an advertisement
*   Train a classification model using user-related features
*   Evaluate model performance using standard metrics


---



⚙️ Methodology
1. Data Preparation
* Load dataset
* Handle missing values (if any)
* Select relevant features
* Define:
X (features),
y (target variable)


---

2. Train-Test Split, Split dataset into:
* Training set → for learning
* Test set → for evaluation
* Ensures proper generalization


---

3. Model Training

A Decision Tree Classifier is trained with:

* max_depth = 3 → prevents overfitting
* random_state = 42 → ensures reproducibility

The model learns rules to classify users based on input features.


---

4. Prediction
* Model predicts labels on unseen test data
* Outputs whether a user will click an ad or not


---

5. Evaluation Metrics

The model is evaluated using:

* Accuracy → overall correctness
* Precision → correctness of positive predictions
* Recall → ability to detect actual positives
* F1-score → balance between precision and recall


---

6. Confusion Matrix

Shows correct vs incorrect predictions

Helps analyze:
* False positives
* False negatives

✅ Advantages

* Easy to understand and interpret
* Requires minimal data preprocessing

❌ Limitations
* Prone to overfitting if not controlled
* Sensitive to small changes in data
* Can create overly complex trees
* Lower accuracy compared to ensemble methods
