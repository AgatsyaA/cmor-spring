📌 Overview

This project demonstrates Principal Component Analysis (PCA), an unsupervised machine learning technique used for dimensionality reduction. PCA transforms high-dimensional data into a smaller set of uncorrelated variables while preserving most of the original variance.

The algorithm is applied to the Wine dataset to reduce 13 features into 2 principal components for visualization and analysis.

PCA (Principal Component Analysis) is a dimensionality reduction technique and helps us to reduce the number of features in a dataset while keeping the most important information. It changes complex datasets by transforming correlated features into a smaller set of uncorrelated components.

It helps us to remove redundancy, improve computational efficiency and make data easier to visualize and analyze.

🧠 What is PCA?

Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms data into a new coordinate system using principal components.

👉 In simple terms:
It reduces the number of features while keeping the most important information.



---
🎯 Objective
* Reduce dimensionality of high-dimensional data
* Preserve maximum variance
* Identify important patterns
* Enable visualization of complex datasets



---
⚙️ Methodology / Concept Explanation

PCA works by:
* Transforming original features into orthogonal (uncorrelated) components
* Each component is a linear combination of original features
* Components are ranked by variance explained



---
📊 Dataset Description

The Wine dataset consists of physicochemical measurements of wines derived from three different cultivars. It includes 13 numerical features such as alcohol content, malic acid, ash, alcalinity of ash, magnesium, total phenols, flavanoids, nonflavanoid phenols, proanthocyanins, color intensity, hue, OD280/OD315, and proline. It also contains a target variable representing wine class, used only for visualization since PCA is unsupervised. The dataset is well-suited for PCA due to its multiple numerical features, correlated variables, and relatively high dimensionality.

⚙️ Implementation Steps

🔹 1. Load Dataset
* Load Wine dataset using sklearn.datasets

🔹 2. Exploratory Analysis
* Check dataset shape
* Review statistical summary
* Understand feature distribution

🔹 3. Feature Scaling

Apply StandardScaler Ensures:
* Mean = 0
* Standard deviation = 1

👉 Important because PCA is variance-based.

🔹 4. Apply PCA
* Fit PCA on standardized data
* Transform features into principal components

🔹 5. Explained Variance Analysis
* Analyze explained variance ratio
* Plot cumulative variance
* Decide number of components

🔹 6. Dimensionality Reduction
* Reduce dataset to 2 principal components

🔹 7. Visualization

* Scatter plot of PC1 vs PC2
* Color-coded by wine class

🔹 8. Feature Contribution (Loadings)
* Analyze which features contribute most to components

✅ Advantages
* Reduces dimensionality effectively
* Removes redundancy (correlated features)
* Improves visualization
* Reduces noise in data
* Speeds up model training

❌ Limitations
* Loss of interpretability (transformed features)
* Information loss during reduction
* Sensitive to scaling
* Assumes linear relationships

📌 Applications
* Data visualization (2D/3D plots)
* Preprocessing for ML models
* Noise reduction
* Image compression
* Feature extraction
